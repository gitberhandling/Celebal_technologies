```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset('titanic')

```

```python
print(titanic.head())

```

```python
sns.countplot(x='survived', data=titanic)
plt.title("Survival Count")
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.show()
```

```python
# Survival by class
sns.countplot(x='class', hue='survived', data=titanic)
plt.title("Survival by Passenger Class")
plt.legend(['Not Survived', 'Survived'])
plt.show()
```

```python
# 4. Age distribution by survival
sns.histplot(data=titanic, x='age', hue='survived', bins=30, kde=True, element='step')
plt.title("Age Distribution by Survival")
plt.show()
```

```python
# Survival based on embark_town
sns.countplot(x='embark_town', hue='survived', data=titanic)
plt.title("Survival by Embarkation Town")
plt.legend(['Not Survived', 'Survived'])
plt.show()
```

```python
#Boxplot: Age vs. Class
sns.boxplot(x='class', y='age', data=titanic)
plt.title("Age Distribution Across Classes")
plt.show()
```

```python
#Pairplot: Selected features
sns.pairplot(titanic[['age', 'fare', 'survived', 'pclass']], hue='survived')
plt.suptitle("Pairplot of Age, Fare, Class and Survival", y=1.02)
plt.show()
```

```python
# Pie Chart: Survival distribution
survived_counts = titanic['survived'].value_counts()
labels = ['Not Survived', 'Survived']
plt.figure(figsize=(6, 6))
plt.pie(survived_counts, labels=labels, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightgreen'])
plt.title('Survival Distribution (Pie Chart)')
plt.axis('equal')
plt.show()
```

2. How the price of the ticket (column name: 'fare') for each passenger
is distributed by plotting a histogram.:

```python
# Plot a histogram to visualize the distribution of 'fare'
plt.figure(figsize=(10, 6))
sns.histplot(titanic['fare'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Titanic Ticket Fares')
plt.xlabel('Fare')
plt.ylabel('Number of Passengers')
plt.grid(True)
plt.show()
```
